3. App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. To bind a function to an URL path we use the app.route decorator.  
Ex-

```python
from flask import Flask
  
app = Flask(__name__)

#This decorator takes the required route  
@app.route("/profile")
def profile():
    return "This is profile route!"  
    
@app.route("/")
def home():
    return "This is home route!"  
  
if __name__ == "__main__":
    app.run(debug=True)
```

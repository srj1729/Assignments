Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. It allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc. Plots that can be plotted using the Pyplot module of Matplotlib are -  
i. Line Plot  
ii. Bar Chart Plot  
iii. Pie Plot  
iv. Scatter Plot  
v. 3D Plot

2. 

Scatter Plot is the dot-based plotting of multiple variables along x and y axes. We can use different colors for different dots if necessary for better plotting and identification of dots.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
```

3. 

subplot() function divides the current figure into rectangular panes that are numbered row-wise. Each pane contains an axes. Subsequent plots are output to the current pane.

```python
import numpy as np
a = np.array([0, 1, 2, 3, 4, 5])
b = np.array([0, 100, 200, 300, 400, 500])
plt.figure()
plt.subplot(2,2,1)
plt.plot(a, b, color="b")
c = np.array([0, 1, 2, 3, 4, 5])
d = np.array([50, 20, 40, 20, 60, 70])
plt.figure()
plt.subplot(2,2,2)
plt.plot(c, d, color="b")
e = np.array([0, 1, 2, 3, 4, 5])
f = np.array([10, 20, 30, 40, 50, 60])
plt.figure()
plt.subplot(2,2,3)
plt.plot(e, f, color="b")
g = np.array([0, 1, 2, 3, 4, 5])
h = np.array([200, 350, 250, 550, 450, 150])
plt.figure()
plt.subplot(2,2,4)
plt.plot(g, h, color="b")
plt.show()
```

4. 

Categorical data can be represented in rectangular blocks with different heights or lengths proportional to the values. Such a type of representation is called a bar chart. Bar charts can be used to plot data in both vertical and horizontal manner.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot')
plt.show()
```

5. 

Box plots are used to show distributions of numeric data values, especially when we want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2],labels=['Box1','Box2'])
plt.show()
```

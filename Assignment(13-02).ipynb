The ArithmeticError exception class serves as a base class for those built-in exceptions that are raised due to various arithmetic errors like OverflowError, ZeroDivisionError, FloatingPointError.  
Ex-

```python
#Example of ZeroDivisionError:
try:
    1/0
except ArithmeticError as e:
    print(e)
```

```python
#Example of OverflowErrorError:
j = 5.0
try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(e)
```

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively. LookupError exception class  can be used to handle both IndexError and KeyError exception classes.  
Ex-

```python
#IndexError
try:
    l=[2,4,234,565,6,75]
    for i in range(0,10):
        print(l[i])
except IndexError as e:
    print(e)
```

As we can see IndexError occurs because i exceeds the limit of indices in list.

```python
#KeyError
dic={"name":"rahul","age":"13","gender":"male"}
try:
    print(dic["name"])
    print(dic["height"])
except KeyError as e:
    print(e)
```

As we can see KeyError occurs because the key height is not present in dictionary.

```python
items=['A','B','C','D','E','F']
values=[3,6,4,8,8,5]
df2=pd.DataFrame({'Items':items,'Values':values})
df2
```

```python
def mean(lst):
    s=0
    for x in lst:
        s=s+x
    return s/len(lst)

def median(lst):
    if len(lst)%2==0:
        return (lst[len(lst)//2]+lst[len(lst)//2-1])/2
    else:
        return lst[len(lst)//2]

def mode(lst):
    d={}
    for x in lst:
        if not x in d:
            d[x]=1
        else:
            d[x]+=1
    
    return [f for f,s in d.items() if s==max(d.values())][0]

print(mean(df2['Values']))
print(median(df2['Values']))
print(mode(df2['Values']))
```

```python
df3=pd.DataFrame({'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05','2023-01-06','2023-01-07','2023-01-08','2023-01-09','2023-01-10','2023-01-11'],
                  'Sales':[310,400,563,381,1096,765,431,561,341,462,891]})
df3                  
```

```python
df3['Date'] = pd.to_datetime(df3.Date)
df3
```

```python
import datetime as dt
def find(x):
    s=0
    l=0
    for y in df3[(df3['Date']<=x) & (df3['Date']>x-dt.timedelta(days=7))]['Sales']:
        s+=y
        l+=1
    return s/l

df3['MovingAverage']=[find(x) for x in df3['Date']]
```

```python
df3
```
